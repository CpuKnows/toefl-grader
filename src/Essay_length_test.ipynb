```python
import pandas as pd
import nltk

essay_key = pd.read_csv('D:/UIC/Spring 2018/NLP/Project/essay_split.csv', sep=',')
essay_key.head()
```

```python
essay_key['grade'].value_counts()
```

```python
essays = []
for filename in essay_key['filename']:
    with open('D:/UIC/Spring 2018/NLP/Project/essays_dataset/essays/'+filename, 'r') as f:
        essays.append(f.read().strip())
        
essay_key['essay'] = essays
essay_key.head()
```

```python
print(essay_key.loc[0,'essay'])
```

```python
sent_text = nltk.sent_tokenize(essay_key.loc[0, 'essay'])
sent_text
```

```python
senteces_length = []
for essay in essay_key['essay']:
        sent_text = nltk.sent_tokenize(essay)
        senteces_length.append(len(sent_text))
        
essay_key['num_senteces'] = senteces_length
essay_key
```
